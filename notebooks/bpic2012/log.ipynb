```python
import pm4py

data = pm4py.read_xes("../../data/bpic2012/origin/BPI_Challenge_2012.xes.gz")
data
```

```python
data = data[["case:concept:name", "concept:name", "time:timestamp"]]
```

```python
from pm4py.objects.log.util.sorting import sort_timestamp_log, sort_timestamp_trace

log = sort_timestamp_log(data)
print(log)
for i, trace in enumerate(log):
    # print first and last event in trace
    print(trace[0]["concept:name"], trace[0]["time:timestamp"], sep="\n")
    print(trace[-1]["concept:name"], trace[-1]["time:timestamp"], sep="\n")

    print()
    if i == 2:
        break
```

```python
log = log[1:]
for i, trace in enumerate(log):
    # print first and last event in trace
    print(trace[0]["concept:name"], trace[0]["time:timestamp"], sep="\n")
    print(trace[-1]["concept:name"], trace[-1]["time:timestamp"], sep="\n")

    print()
    if i == 2:
        break
```
