# BPIC2012

Plot the training curves for the files `history.csv` under the directories `<root>/runs/<run>/`

```python
import pathlib

history_files = [
    file for file in pathlib.Path("../../runs/bpic2012").rglob("history.csv")
]
for history_file in history_files:
    print(history_file)
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

for history_file in history_files:

    history = pd.read_csv(history_file)
    for key in history.keys():
        if key == "epoch":
            continue
        plt.plot(
            history["epoch"],
            history[key],
            label=key,
        )

    # plt.plot(
    #     history["epoch"],
    #     history["train_loss"],
    #     label=history_file.parent.name + " train loss",
    # )
    # plt.plot(
    #     history["epoch"],
    #     history["val_loss"],
    #     label=history_file.parent.name + " val loss",
    # )
    plt.legend(bbox_to_anchor=(1, 0), loc="lower left")
    plt.show()
```
